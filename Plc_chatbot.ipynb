```python
!pip install -qqq langchain-community langchain-text-splitters langchain-huggingface
from langchain_community.document_loaders import TextLoader
loadder  = TextLoader('plc.txt',encoding="utf-8")
loadder
```

```python
%%writefile plc.txt
This is a sample PLC text file.
It contains information about Programmable Logic Controllers.
PLCs are used for industrial automation.
```

```python
text_documents = loadder.load()
text_documents
```

```python
!pip install -qqq langchain-text-splitters
from langchain_text_splitters import RecursiveCharacterTextSplitter
textSplitter = RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=20)
finalDocuments = textSplitter.split_documents(text_documents)
finalDocuments
```

```python
texts = [doc.page_content for doc in finalDocuments]
```

```python
!pip install -qqq langchain-huggingface
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
print(embeddings)
```

```python
!pip install -qqq faiss-cpu
from langchain_community.vectorstores import FAISS
db = FAISS.from_documents(finalDocuments, embeddings)
db
```

```python
query = input("Enter your question for PLC : ")
docs = db.similarity_search(query)
for i in range(len(docs)):
    print(docs[i].page_content)
```
